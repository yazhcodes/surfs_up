```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_result = session\
                    .query(Measurement.tobs)\
                    .filter(func.extract("month",Measurement.date) == 6)\
                    .all()
june_result
```

```python
#  3. Convert the June temperatures to a list.
june_temps = list(np.ravel(june_result))
# june_temps = [j for i in june_result for j in i] # alternate method to disintegrate a list of tuples into a list.
june_temps
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
# Creating DataFrame using Query Result(june_result) instead of the List(june_temps) 
# from Step 3 as it produces an incorrect DataFrame
june_temps_df = pd.DataFrame(june_result,columns=['June Temps'])
june_temps_df.head()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_result = session.query(Measurement.tobs).filter(func.extract("month",Measurement.date) == 12).all()
dec_result
```

```python
# 7. Convert the December temperatures to a list.
dec_temps = list(np.ravel(dec_result))
dec_temps
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
# Creating DataFrame using Query Result(dec_result) instead of the List(dec_temps) 
# from Step 7 as it produces an incorrect DataFrame
dec_temps_df = pd.DataFrame(dec_result,columns=['December Temps'])
dec_temps_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temps_df.describe()
```

## ADDITIONAL QUERIES: Including Precipitation and Stations in the Analysis

```python
months = {
    '1':'January',
    '2':'February',
    '3':'March',
    '4':'April',
    '5':'May',
    '6':'June',
    '7':'July',
    '8':'August',
    '9':'September',
    '10':'October',
    '11':'November',
    '12':'December'
}
```

```python
# Including Precipitation analysis
def temp_prcp_analysis(month):
    month_name = months[f'{month}']
    result = session.query(Measurement.tobs,Measurement.prcp)\
            .filter(func.extract("month",Measurement.date) == month).all()
    df = pd.DataFrame(result,columns=[f'{month_name} Temperature',f'{month_name} Precipitation'])
    return df

jun_df = temp_prcp_analysis(6)
dec_df = temp_prcp_analysis(12)
```

```python
jun_df.describe()
```

```python
dec_df.describe()
```

```python
# Station Wise Monthly Averages
def station_wise_avg(month):
    month_name = months[f'{month}']
    result = session.query(
                        Measurement.station,
                        func.round(func.avg(Measurement.tobs),2),
                        func.round(func.avg(Measurement.prcp),2)
                        )\
                    .filter(func.extract("month",Measurement.date) == month)\
                    .group_by(Measurement.station)\
                    .all()
    df = pd.DataFrame(result,columns=['Stations',f'{month_name} Temperature',f'{month_name} Precipitation'])
    return df

jun_df = station_wise_avg(6)
dec_df = station_wise_avg(12)
```
